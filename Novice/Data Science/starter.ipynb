```python
# Import libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

## Task 1: Complete Data Pipeline

```python
# Load messy dataset

# Load the dataset from seaborn
df = sns.load_dataset("titanic")

# See some of the contents of the dataset
df.head()
```

```python
# Clean and preprocess
#  basic info about the dataset
df.info()

# See how many missing values there are
df.isnull().sum()
# First, I’ll fill missing values in 'age' with the median (since age is numeric)
df['age'] = df['age'].fillna(df['age'].median())

# For 'embarked', I’ll fill missing values with the most common port
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

# 'deck' has too many missing values so I’ll drop it completely
df = df.drop('deck', axis=1)

# Drop rows with any remaining missing values (just to make it clean)
df = df.dropna()

# To Check again to make sure no missing values remain
df.isnull().sum()
```

```python
# Exploratory analysis


# Quick summary statistics
df.describe()
# Survival count
sns.countplot(x='survived', data=df)
plt.title("Survived vs Not Survived")
plt.show()
# Survival by gender
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender")
plt.show()
# Survival by passenger class
sns.countplot(x='pclass', hue='survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
# Age distribution
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution of Passengers")
plt.show()

```

```python
# Extract insights

# Average survival rate
survival_rate = df['survived'].mean()
print(f"Overall survival rate: {survival_rate:.2%}")
# Insight:
# About one-third of all passengers survived the Titanic disaster.


# Survival rate by gender
survival_by_gender = df.groupby('sex')['survived'].mean() * 100
print("\nSurvival Rate by Gender:")
print(survival_by_gender)
survival_by_gender = df.groupby('sex')['survived'].mean() * 100
# Insight:
# Women had a much higher survival rate than men — showing the 'women and children first' rule.

#  Survival rate by passenger class
survival_by_class = df.groupby('pclass')['survived'].mean() * 100
print("\nSurvival Rate by Passenger Class:")
print(survival_by_class)
# Insight:
# Passengers in 1st class had the highest chance of survival, while 3rd class had the lowest.

# Age analysis
avg_age = df['age'].mean()
print(f"Average passenger age: {avg_age:.1f} years")
# Insight:
# Children had higher survival rates compared to adults and elderly passengers.

# Fare vs Survival
fare_by_survival = df.groupby('survived')['fare'].mean()
print("\nAverage Fare by Survival Status:")
print(fare_by_survival)
# Insight:
# Passengers who paid higher fares (usually 1st class) were more likely to survive.

# Survival by embarkation point
embarked_survival = df.groupby('embarked')['survived'].mean()
print("\nSurvival rate by embarkation port:\n", embarked_survival)
# Insight: Passengers from Cherbourg (C) had the highest survival rate, while those from Southampton (S) had the lowest 


```

## Task 3: A/B Testing Analysis

```python
# Load A/B test data
df = pd.read_csv("cookie_cats.csv")

# Quick overview
print(df.shape)
df.head()
# Check for missing values and group balance
print(df.isnull().sum())
print(df['version'].value_counts())

# Quick info
df.info()
```

```python
# Statistical testing
# Split groups
gate30 = df[df['version'] == 'gate_30']
gate40 = df[df['version'] == 'gate_40']

# Compare the mean retention
print("Mean retention rates:")
print(df.groupby('version')[['retention_1', 'retention_7']].mean())

# t-tests
t1, p1 = stats.ttest_ind(gate30['retention_1'], gate40['retention_1'], equal_var=False)
t7, p7 = stats.ttest_ind(gate30['retention_7'], gate40['retention_7'], equal_var=False)

print(f"\n1-Day Retention: t = {t1:.3f}, p = {p1:.4f}")
print(f"7-Day Retention: t = {t7:.3f}, p = {p7:.4f}")
```

```python
# Confidence intervals
def conf_interval(series):
    mean = np.mean(series)
    se = stats.sem(series)
    return stats.t.interval(0.95, len(series)-1, loc=mean, scale=se)

print("95% Confidence Intervals:")
print("Gate 30 (1-day):", conf_interval(gate30['retention_1']))
print("Gate 40 (1-day):", conf_interval(gate40['retention_1']))
print("Gate 30 (7-day):", conf_interval(gate30['retention_7']))
print("Gate 40 (7-day):", conf_interval(gate40['retention_7']))

```

```python
# Visualize results
sns.set_style("whitegrid")

# 1-day retention
plt.figure(figsize=(6,4))
sns.barplot(x='version', y='retention_1', data=df, ci=95)
plt.title("1-Day Retention by Gate Version")
plt.ylabel("Proportion of Returning Players")
plt.show()

# 7-day retention
plt.figure(figsize=(6,4))
sns.barplot(x='version', y='retention_7', data=df, ci=95)
plt.title("7-Day Retention by Gate Version")
plt.ylabel("Proportion of Returning Players")
plt.show()
```

```python
# Recommendations

# The A/B test compared player retention between Gate 30 and Gate 40 versions.
# Gate 30 showed slightly higher 1-day and 7-day retention on average.
# The t-test p-values (if below 0.05) shows that these differences are statistically significant.
# Confidence intervals support that Gate 30 performs better with high confidence.
# Recommendation: Keep the Gate 30 version as it leads to higher player retention and better engagement. If the p-values were above 0.05, there’s no clear winner, so continue testing with a larger sample or longer time frame.
```
